```python
cls = df.columns
for i in cls:
    if df[i].isnull().sum() != 0:
        print(i,'=',df[i].isnull().sum())
```

```python
df.shape
```

```python
#deleing the columns
del_col = ['Id','Alley','FireplaceQu','PoolQC','Fence','MiscFeature']
```

```python
for i in del_col:
    df = df.drop([i],axis=1)
```

```python
cls = df.columns
fill_col = []
for i in cls:
    if df[i].isnull().sum() != 0:
        print(i,'=',df[i].isnull().sum())
        fill_col.append(i)
```

```python
for i in fill_col:
    print(i ,':',df[i].dtype)
```

```python
values = ['LotFrontage','MasVnrArea','GarageYrBlt']
```

```python
for j in values:
    df[j] = df[j].fillna(df[j].mean())
    fill_col.remove(j)
```

```python
fill_col
```

```python
for i in fill_col:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
#taking categorical column
cat_col = []
for i in df.columns:
    if df[i].dtype == 'O':
        cat_col.append(i)
```

```python
len(cat_col)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
def cat_into_label(column):
    for i in column:
        df[i] = le.fit_transform(df[i])
    return df
```

```python
cat_into_label(cat_col)
```
